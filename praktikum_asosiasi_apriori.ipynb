```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
grocery_data = pd.read_csv('grocery.csv',header=None)
```

```python
grocery_data
```

```python
grocery_data.shape
```

```python
records = []
for i in range(0,22):
  records.append([str(grocery_data.values[i,j]) for j in range(0,13)])
```

```python
association_rules = apriori(records, min_support = 0.50, min_confidence = 0.5, min_lift = 0.3, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```
