```python
import pandas as pd
import numpy as np

mahasiswa = pd. read_csv('student_scores.csv')
mahasiswa
```

```python
mahasiswa = mahasiswa.drop(['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7'],axis=1)
```

```python
mahasiswa
```

```python
mahasiswa.isnull().sum()
```

```python
mahasiswa['Scores'].mean()
```

```python
mahasiswa.fillna(mahasiswa.mean(), inplace=True)
```

```python
mahasiswa
```

```python
mahasiswa.isnull().sum()
```

```python
mahasiswa.info()
```

```python
x_train = np.array(mahasiswa[['Hours','Scores']])
y_train = np.array(mahasiswa['Grade'])

print(f'x_train : \n{x_train} \n')
print(f'y_train : {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
K = 3

model = KNeighborsClassifier(n_neighbors = K)
model.fit(x_train, y_train)
```

```python
Hours = 8.2
Scores = 77 

x_new = np.array([Hours, Scores]).reshape(1, -1)
x_new
```

```python
y_new = model.predict(x_new)
y_new
```

```python
score_manual = np.array([Hours, Scores])
score_manual
```

```python
x_train
```

```python
from scipy.spatial.distance import euclidean

jarak_nilai = [euclidean(score_manual, d) for d in x_train]
jarak_nilai
```

```python
mahasiswa['euclidean'] = jarak_nilai
mahasiswa.sort_values(['euclidean'])
```

```python
x_test = np.array([[2.5, 21],[5.1, 50],[3.2, 27],[8.5, 75]])
y_test = lb.transform(np.array(['E','D','E','C']))

print(f'x_test : \n {x_test} \n')
print(f'y_test : \n {y_test} \n')
```

```python
y_pred = model.predict (x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score (y_test, y_pred)

print (f'Accuracy: {acc}')
```
